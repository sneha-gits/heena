```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
print(x)
```

```python
plt.scatter(x,y,c='g')
plt.show()
```

```python
plt.scatter(x,y,c='g')

plt.xlabel('x san')
plt.ylabel('y san')
plt.title('Scaterr plot')
```

```python
plt.plot(x,y,c='b',linestyle ='dashed')
plt.show()
```
